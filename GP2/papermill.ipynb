```python
ANCESTRIES = ['AAC', 'AFR', 'AJ', 'AMR', 'CAS', 'EAS', 'EUR', 'MDE', 'SAS']
```

## Run for WGS R11

```python
# Home directory
HOME = "/home/jupyter/"
os.chdir(HOME)
GENE = "ITSN1"
CHROM = str(21)
STARTBP = str(33542501)
ENDBP = str(33999861)
RELEASES = [str(11)]
DT = "wgs"
# Notebook to run
notebook = f"{HOME}/ITSN1_Burden_GP2_R11.ipynb"
```

```python
pm.inspect_notebook(notebook)
```

```python
for RELEASE in RELEASES:
    os.makedirs(f"{HOME}/R{RELEASE}", exist_ok=True)
    os.makedirs(f"{HOME}/R{RELEASE}_{DT}", exist_ok=True)
    for ANC in ANCESTRIES:
        try:
            print(f"> {ANC}\n> Release {RELEASE}")
            pm.execute_notebook(
                input_path=notebook,
                output_path=f"{HOME}/R{RELEASE}/{ANC}_{GENE}_{DT.upper()}.ipynb",
                parameters={
                    "ANC": ANC,
                    "GENE": GENE,
                    "CHROM": CHROM,
                    "STARTBP": STARTBP,
                    "ENDBP": ENDBP,
                    "RELEASE": RELEASE,
                    "DT": DT
                },
                progress_bar=True
            )
        except Exception as e:
            print("Error:", e)
```
